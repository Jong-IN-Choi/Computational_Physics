# 큰 수의 법칙, 중심극한정리 (20240319 49쪽)


주어진 확률밀도함수: $ P(y) = e^{-x}$

변환 함수: $\quad u(x) = 1 $

Monte Carlo 방법을 사용하여 주어진 확률밀도함수를 따르는 변수 $ y $를 추출, 히스토그램으로 표현했다.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# P(y) = e^-x, u(x) = 1

N = 10000000

r = np.random.uniform(0, 1, N)
plt.hist(-np.log(r), bins = 1000)
```

```python
a = r.reshape(1000, 10000)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sum1 = np.sum(a, axis = 1)
plt.hist(sum1, bins = 100)
print('평균 :', np.mean(sum1), '표준편차 : ', np.std(sum1))

plt.subplot(1, 2, 2)
sum2 = np.sum(a, axis = 1)[:100]
plt.hist(sum2, bins = 100)
print('평균 :', np.mean(sum2), '표준편차 : ', np.std(sum2))
```
